200個体から100個体をランダムに選択したいので、random.randitを用いた、整数値しかとらない乱数で100個体をランダムに選んだ。

選んだ個体をtemp_new_geneリストに保存する。

この動作を100回繰り返して世代を更新する。

今回は、この乱数で個体を選ぶとこまで実装した。

**今後何を実装する予定か:**

今回の実装状況だと、100個体を各ループごとに選ぶ際、選んだ個体を元のリストから削除していない。

既に選んだ個体を元のリストから削除して、次のループを正常に実行させるためのプログラムを実装する。

具体的には、以下のように実装していく。

１．乱数を振って、個体を選びtemp_new_listにコピーしてから、元のリストからその個体を削除する。

２．乱数の範囲を[0,199]から[0,199-1]に狭める。

３.２の乱数の範囲で再度１を実行する。

これを100回繰り返して、トーナメント戦を実装する。
